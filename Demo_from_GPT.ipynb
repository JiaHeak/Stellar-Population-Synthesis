```python
import numpy as np
import matplotlib.pyplot as plt

# Synthesizer imports (adjust if API differs slightly)
from synthesizer.particle import Stars, Gas, BlackHoles, Galaxy

from synthesizer import Sed
from unyt import Myr, Msun
```

```python
# Define a simple stellar population
N = 100
ages = np.random.rand(N) * 100 * Myr
metallicities = np.random.rand(N) / 100
initial_masses = np.ones(N) * 1e6 * Msun

stars = Stars(
    initial_masses=initial_masses,
    ages=ages,
    metallicities=metallicities,
)

# Create galaxy
galaxy = Galaxy(stars=stars)

# Generate spectrum
sed = galaxy.get_sed()

# Plot
plt.figure(figsize=(8,5))
plt.loglog(sed.wavelength, sed.luminosity)
plt.xlabel("Wavelength (Angstrom)")
plt.ylabel("Luminosity")
plt.title("Galaxy Spectrum (Simple SSP)")
plt.show()
```
